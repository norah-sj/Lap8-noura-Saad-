```python
import os

main_folder = "lab4_data"

if not os.path.exists(main_folder):
    os.mkdir(main_folder)

folders = ["input", "output", "logs"]

for folder in folders:
    path = os.path.join(main_folder, folder)
    if not os.path.exists(path):
        os.mkdir(path)

for item in os.listdir(main_folder):
    path = os.path.join(main_folder, item)
    if os.path.isfile(path):
        print(item, "is a file")
    else:
        print(item, "is a folder")

old_name = os.path.join(main_folder, "logs")
new_name = os.path.join(main_folder, "history")

if os.path.exists(old_name):
    os.rename(old_name, new_name)
```

```python
file_name = "lab4_data/input/notes.txt"

with open(file_name, "w") as f:
    f.write("This is my notes file\n")
    f.write("I am learning Python\n")
    f.write("File handling is easy\n")
    f.write("This is Lab 4 task\n")
    f.write("End of the file\n")

with open(file_name, "r") as f:
    lines = f.readlines()
    print("Number of lines:", len(lines))

copy_file = "lab4_data/output/summary.txt"

with open(file_name, "r") as source, open(copy_file, "w") as target:
    target.write(source.read())
```
